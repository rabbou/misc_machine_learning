## *Random Forest and Adaboost for MNIST*

```python
import numpy as np 
import random
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from matplotlib.ticker import PercentFormatter

data = np.float64(np.load('/project2/cmsc25025/mnist/MNIST.npy'))
numbers = np.float32(np.load('/project2/cmsc25025/mnist/MNIST_labels.npy'))
L = len(data)
training = data[:int(.6*L)]
validation = data[int(.6*L):int(.8*L)]
testing = data[int(.8*L):]
training_num = numbers[:int(.6*L)]
validation_num = numbers[int(.6*L):int(.8*L)]
testing_num = numbers[int(.8*L):]
```

## (a)

```python
def plot_error(train, val):
    n = len(train)+1
    plt.plot(range(1, n),train, 'b', label="training")
    plt.plot(range(1, n),val, 'r', label="validation")
    plt.xticks(range(1, n))
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    plt.xlabel("# of trees")
    plt.ylabel("Error")
    plt.title("Error vs. Number of Trees")
    plt.legend()
    plt.show()
```

```python
def test_tree(min_samples_split, max_features, num_trees):
    errors_train = []
    errors_val = []
    for num in range(1, num_trees+1):
        clf = RandomForestClassifier(n_estimators=num, 
                 min_samples_split=min_samples_split, 
                 criterion="entropy",max_features=max_features)
        clf.fit(training, training_num)
        pred_train = clf.predict(training)
        errors_train.append(np.mean(pred_train != training_num))
        clf.predict_proba(validation)
        pred_val = clf.predict(validation)
        errors_val.append(np.mean(pred_val != validation_num))
    best_train_e = min(errors_train)*100
    best_train_i = np.argmin(errors_train) + 1
    best_val_e = min(errors_val)*100
    best_val_i = np.argmin(errors_val) + 1
    print("best training error: {:.2f}% obtained with {} tree(s)" \
          .format(best_train_e, best_train_i))
    print("best validation error: {:.2f}% obtained with {} tree(s)" \
          .format(best_val_e, best_val_i))
    return errors_train, errors_val
```

```python
errors_train, errors_val = test_tree(3, 3, 15)
plot_error(errors_train, errors_val)
```

```python
errors_train, errors_val = test_tree(2, 10, 15)
plot_error(errors_train, errors_val)
```

```python
errors_train, errors_val = test_tree(3, 20, 15)
plot_error(errors_train, errors_val)
```

```python
errors_train, errors_val = test_tree(2, 20, 15)
plot_error(errors_train, errors_val)
```

```python
errors_train, errors_val = test_tree(2, 50, 15)
plot_error(errors_train, errors_val)
```

It seems like a lower minimum number of samples improves the error rate. Similarly, as max_features increase, the error decreases. Let's look at a plot of the error vs each parameter, fixing the number of trees to 20.

```python
errors_train = []
errors_val = []
for split in range(2, 11):
    clf = RandomForestClassifier(n_estimators=20, 
             min_samples_split=split, 
             criterion="entropy",max_features=50)
    clf.fit(training, training_num)
    pred_train = clf.predict(training)
    errors_train.append(np.mean(pred_train != training_num))
    clf.predict_proba(validation)
    pred_val = clf.predict(validation)
    errors_val.append(np.mean(pred_val != validation_num))
best_train_e = min(errors_train)*100
best_train_i = np.argmin(errors_train) + 2
best_val_e = min(errors_val)*100
best_val_i = np.argmin(errors_val) + 2
print("best training error: {:.2f}% obtained with min_sample_split = {}" \
      .format(best_train_e, best_train_i))
print("best validation error: {:.2f}% obtained with min_sample_split = {}" \
      .format(best_val_e, best_val_i))
```

```python
n = len(errors_train)+1
plt.plot(range(2, 11), errors_train, 'b', label="training")
plt.plot(range(2, 11), errors_val, 'r', label="validation")
plt.xticks(range(2, 11))
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.xlabel("Minimum samples split")
plt.ylabel("Error")
plt.title("Error vs. Min samples split")
plt.legend()
plt.show()
```

2 samples split seem to lead to the best validation error rate.

```python
errors_train = []
errors_val = []
for feat in range(5, 100, 5):
    clf = RandomForestClassifier(n_estimators=20, 
             min_samples_split=3, 
             criterion="entropy",max_features=feat)
    clf.fit(training, training_num)
    pred_train = clf.predict(training)
    errors_train.append(np.mean(pred_train != training_num))
    clf.predict_proba(validation)
    pred_val = clf.predict(validation)
    errors_val.append(np.mean(pred_val != validation_num))
best_train_e = min(errors_train)*100
best_train_i = range(5, 100, 5)[np.argmin(errors_train)]
best_val_e = min(errors_val)*100
best_val_i = range(5, 100, 5)[np.argmin(errors_val)]
print("best training error: {:.2f}% obtained with max_features = {}" \
      .format(best_train_e, best_train_i))
print("best validation error: {:.2f}% obtained with max_features = {}" \
      .format(best_val_e, best_val_i))
```

```python
n = len(errors_train)+1
plt.plot(range(5, 100, 5), errors_train, 'b', label="training")
plt.plot(range(5, 100, 5), errors_val, 'r', label="validation")
plt.xticks(range(5, 100, 5))
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.xlabel("Max features")
plt.ylabel("Error")
plt.title("Error vs. Max features")
plt.legend()
plt.show()
```

Therefore, our best classifier has parameters:
 - num_trees: 20
 - min_samples_split: 3
 - max_features: 60

```python
clf = RandomForestClassifier(n_estimators=20, 
         min_samples_split=3, 
         criterion="entropy",max_features=60)
clf.fit(training, training_num)
clf.predict_proba(testing)
pred_test = clf.predict(testing)
print("error on testing set: {:.2f}%"\
      .format(100*np.mean(pred_test != testing_num)))
```

This protocol performs pretty well on this test set, as we obtained a low error rate.

## (b)

```python
def test_tree_clfb(min_samples_split, max_features, num_trees, clfb_trees):
    errors_train = []
    errors_val = []
    for num in range(1, num_trees+1):
        clfb = RandomForestClassifier(n_estimators=clfb_trees,
                      min_samples_split=min_samples_split, 
                     criterion="entropy",max_features=max_features)
        clf = AdaBoostClassifier(clfb, n_estimators=num,
                                 algorithm='SAMME')
        clf.fit(training, training_num)
        pred_train = clf.predict(training)
        errors_train.append(np.mean(pred_train != training_num))
        clf.predict_proba(validation)
        pred_val = clf.predict(validation)
        errors_val.append(np.mean(pred_val != validation_num))
    best_train_e = min(errors_train)*100
    best_train_i = np.argmin(errors_train) + 1
    best_val_e = min(errors_val)*100
    best_val_i = np.argmin(errors_val) + 1
    print("best training error: {:.2f}% obtained with {} tree(s)" \
          .format(best_train_e, best_train_i))
    print("best validation error: {:.2f}% obtained with {} tree(s)" \
          .format(best_val_e, best_val_i))
    return errors_train, errors_val
```

Let's first look at the error rates.

```python
errors_train, errors_val = test_tree_clfb(3, 60, 20, 1)
plot_error(errors_train, errors_val)
```

The AdaBoost classifier seems to be doing pretty well only using 1 tree.

```python
errors_train, errors_val = test_tree_clfb(3, 60, 10, 5)
plot_error(errors_train, errors_val)
```

```python
def test_tree_clfb2(min_samples_split, max_features, num_trees, clfb_trees):
    errors_train = []
    errors_val = []
    for num in range(1, clfb_trees+1):
        clfb = RandomForestClassifier(n_estimators=num,
                      min_samples_split=min_samples_split, 
                     criterion="entropy",max_features=max_features)
        clf = AdaBoostClassifier(clfb, n_estimators=num_trees,
                                 algorithm='SAMME')
        clf.fit(training, training_num)
        pred_train = clf.predict(training)
        errors_train.append(np.mean(pred_train != training_num))
        clf.predict_proba(validation)
        pred_val = clf.predict(validation)
        errors_val.append(np.mean(pred_val != validation_num))
    best_train_e = min(errors_train)*100
    best_train_i = np.argmin(errors_train) + 1
    best_val_e = min(errors_val)*100
    best_val_i = np.argmin(errors_val) + 1
    print("best training error: {:.2f}% obtained with {} tree(s)" \
          .format(best_train_e, best_train_i))
    print("best validation error: {:.2f}% obtained with {} tree(s)" \
          .format(best_val_e, best_val_i))
    return errors_train, errors_val
```

```python
errors_train, errors_val = test_tree_clfb(3, 5, 10, 5)
plot_error(errors_train, errors_val)
```

The plot above shows this time the error versus the number of trees selected in clfb. Since the error rate gets better as we increased the number of trees of the clfb, let's fix the parameters and study the error rate.

```python
clfb = RandomForestClassifier(n_estimators=20,
              min_samples_split=3, 
             criterion="entropy",max_features=60)
clf = AdaBoostClassifier(clfb, n_estimators=20,
                         algorithm='SAMME')
clf.fit(training, training_num)
pred_train = clf.predict(training)
print('training error : {:.2f}%'.format(100*np.mean(pred_train != training_num)))
clf.predict_proba(validation)
pred_val = clf.predict(validation)
print('validation error : {:.2f}%'.format(100*np.mean(pred_val != validation_num)))
```

The error rate is definitely better as the number of trees for clfb classifier increases. Let's try on the test set.

```python
clf.predict_proba(testing)
pred_test = clf.predict(testing)
print("error on testing set: {:.2f}%"\
      .format(100*np.mean(pred_test != testing_num)))
```

We obtained a better classification error on the test set than in (a).
